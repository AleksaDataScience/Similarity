```python
# Assuming you have a DataFrame named df with the specified columns
features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement", "view", "bathrooms",
            "sqft_living15", "sqft_above", "grade", "sqft_living"]

X = df[features]  # Select the list of features as the feature matrix
Y = df['price']   # Target variable 'price'

# Create a list of tuples with estimator names and model constructors
estimators = [
    ('scale', StandardScaler()),                  # Scaling
    ('polynomial', PolynomialFeatures(include_bias=False)),  # Polynomial Features
    ('model', LinearRegression())                # Linear Regression
]

# Create a pipeline object
pipeline = Pipeline(estimators)

# Fit the pipeline with the selected features and target variable
pipeline.fit(X, Y)

# Calculate the R-squared (R^2) value
r_squared = pipeline.score(X, Y)

# Print the R-squared value
print("R-squared (R^2) value:", r_squared)
```
